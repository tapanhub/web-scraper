```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
olxmobilelink='https://www.olx.in/bangalore/mobile-phones/?search%5Border%5D=filter_float_price%3Aasc'

def getdata(link=olxmobilelink):
    
    data=[]
    
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')

    tbls = soup.find_all('table', attrs={'class':'fixed'})
    for tbl in tbls:
        price=tbl.find('strong', attrs={'class':'c000'})
        if price:
            mobileinfo={}
            #print(price.text)
            bnd=tbl.find('small', attrs={'class':'breadcrumb'})
            if bnd:
                info=bnd.text.split('\n')
                brand=info[1].split('Â»')[1].replace(' ', '')
                area=info[-1].split(',')[0]
                city=info[-1].split(',')[-1]
                mobileinfo['brand']=brand
                mobileinfo['area']=area
                mobileinfo['city']=city
                mobileinfo['price']=int(price.text)
                #print (info)
                #print((brand, area, city))
                #print (brand.text)
            h3=tbl.find('h3')
            if h3:
                add=h3.find('span')
                if add:
                    description=add.text
                    mobileinfo['desc']=description
                    data.append(mobileinfo)
    return data


#https://www.olx.in/bangalore/mobile-phones/?search%5Border%5D=filter_float_price%3Aasc
#https://www.olx.in/bangalore/mobile-phones/?search%5Border%5D=filter_float_price%3Aasc&page=2
data=[]
for i in range(1, 2):
    data=getdata("https://www.olx.in/bangalore/mobile-phones/?search%%5Border%%5D=filter_float_price%%3Aasc&page=%d"%(i))
    data = data + data
    time.sleep(2)
    


df=pd.DataFrame(data)
column_order=['brand', 'price', 'desc', 'area', 'city']
df=df[column_order]
print(df)
```
